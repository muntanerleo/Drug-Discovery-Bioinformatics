### Preparing the X and Y Data Matrices

```python
# TODO: Create X data matrix: 
# * The X descriptors will be the molecular descriptors which are the Pubchem fingerprints
df3_X = pd.read_csv('data/descriptors_output.csv')
df3_X
```

```python
# Since I only want the molecular features, I will dorp the Name column 
df3_X = df3_X.drop(columns=['Name'])
df3_X
```

```python
# TODO: Create Y data matrix

# * Convert EC50 to pEC50:
df3_Y = df3['pEC50']
df3_Y
```

```python
# TODO: Combining X and Y variables 

# * I will combine these and output it into a csv file which i will be using for model building in part 4. 
dataset3 = pd.concat([df3_X,df3_Y], axis=1)
dataset3
```
